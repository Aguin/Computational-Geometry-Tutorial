```python
import numpy as np
import matplotlib.pyplot as plt
```

# 直线求交

```python
def line_inter(A, B, C, D):
    return A + (B - A) * np.cross(C - A, D - C) / np.cross(B - A, D - C)
```

```python
pts = np.random.randn(4, 2)
P = line_inter(pts[0], pts[1], pts[2], pts[3])

plt.scatter(pts[:2, 0], pts[:2, 1], c="r")
plt.axline(pts[0], pts[1], c="r")

plt.scatter(pts[2:, 0], pts[2:, 1], c="g")
plt.axline(pts[2], pts[3], c="g")

plt.scatter(P[0], P[1], c="b")

plt.show()
```

# 点云可视化

```python
pts = np.fromfile("./data/pa2_Pedestrian.bin", dtype=np.float32).reshape(-1, 4)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.axis("off")
ax.scatter(pts[:, 0], pts[:, 1], pts[:, 2])
ax.set_box_aspect([ub - lb for lb, ub in (getattr(ax, f"get_{a}lim")() for a in "xyz")])
```

```python
pts = np.fromfile("./data/pa2_Cyclist.bin", dtype=np.float32).reshape(-1, 4)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.axis("off")
ax.view_init(30, 100)
ax.scatter(pts[:, 0], pts[:, 1], pts[:, 2], color=plt.get_cmap("Set2")(5))
ax.set_box_aspect([ub - lb for lb, ub in (getattr(ax, f"get_{a}lim")() for a in "xyz")])
```
